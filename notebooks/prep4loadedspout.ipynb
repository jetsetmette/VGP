```python
from pathlib import Path
import pandas as pd
```

```python
file = "C:/Github/VGP/data/Events2023-08-28T14_09_07.csv"

data = pd.read_csv(Path(file))
```

```python
event_rows = []
for row in data.iterrows():
    event = row[1][0]
    if event == "pump1_on" or event == "pump2_on":
        try:
            #write current data, if none then skip
            event_rows.append([current_event, current_start_frame, nlicks1, nlicks2])
        except:
            pass
        
        current_event = event
        current_start_frame = row[1][15]
        nlicks1 = 0
        nlicks2 = 0
    elif event == "Lick1":
        nlicks1 += 1
    elif event == "Lick2":
        nlicks2 += 1
```

```python
cols = ["event", "start_frame", "nlicks1", "nlicks2"]

df = pd.DataFrame(event_rows, columns=cols)
```

```python
def make_loaded_column(df, event, lickscol):
    loaded = False
    newcol = []
    for row in df.iterrows():
        newcol.append(loaded)
        if row[1][0] == event and row[1][lickscol] == 0:
            loaded = True
        elif row[1][lickscol] > 0:
            loaded = False
    return newcol

df['spout1_loaded'] = make_loaded_column(df, 'pump1_on', 2)
df['spout2_loaded'] = make_loaded_column(df, 'pump2_on', 3)
```
